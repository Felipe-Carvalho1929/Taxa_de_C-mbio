```python
moeda = input("Digite o nome da moeda: ")

# Função para pesquisar o código pela moeda
def buscar_codigo_por_moeda(moeda):
    moedas_encontradas = df[df['Moeda'].str.lower() == moeda.lower()]
    codigos = moedas_encontradas['Código'].tolist()
    return codigos[0] if codigos else None

# Buscar o código da moeda informada
codigo = buscar_codigo_por_moeda(moeda)

# Verificar se a moeda foi encontrada
if codigo:
    print("Código da moeda {}: {}".format(moeda, codigo))
else:
    print("Moeda não encontrada.")
```

```python
def obter_taxa_cambio(moeda_base, moeda_alvo):
    try:
        url = f"https://api.exchangerate-api.com/v4/latest/{moeda_base}"
        response = requests.get(url)
        response.raise_for_status()  
        dados = response.json()
        if moeda_alvo in dados['rates']:
            taxa = dados['rates'][moeda_alvo]
            return taxa
        else:
            print("Moeda alvo inválida")
    except requests.exceptions.RequestException as e:
        print(f"Erro na solicitação HTTP: {e}")
    except KeyError:
        print("Moeda alvo inválida ou dados de taxa de câmbio ausentes")

moeda_base = codigo  
moeda_alvo = 'BRL'  # Real brasileiro

taxa_cambio = obter_taxa_cambio(moeda_base, moeda_alvo)
if taxa_cambio is not None:
    taxa_cambio = round(taxa_cambio, 2)
    print(f"A taxa de câmbio de {moeda_base} para {moeda_alvo} é {taxa_cambio}")
```
